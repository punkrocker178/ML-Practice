```python
import numpy as np
import matplotlib.pyplot as plt
X = [1,2,3,4,5,6,7]
y = [2.05,4,6.1,8.7,10,11.5,14.1]
print(X)
print(y)
```

```python
# step 1: initialize and forward propagate  
# y = w_1 * x + w_0
W = [1,3]
def f(x):
    return W[1]*x+W[0] 

print(f(1)) 
```

```python
# step 2: loss function 
def loss():
    sum = 0
    for i in range(len(X)):
        sum += (f(X[i])-y[i])**2
    return sum/(2*len(X)) 
print(loss())
```

```python
# step 3: Differentiation
def derivate():
    sum1 = 0
    sum0 = 0
    for i in range(len(X)):
        sum1 += (f(X[i])-y[i])*X[i]
        sum0 += (f(X[i])-y[i])
    return sum0/len(X),sum1/len(X)

print(derivate())
```

```python
# step 4: update weight 
W = [1,13]
epoch = 20
learning_rate = 0.02
epsilon = 0.0001
for i in range(epoch):
    before_loss = loss()
    d0,d1=derivate()
    print(W)
    W[1] = W[1] - learning_rate*d1
    W[0] = W[0] - learning_rate*d0
    print(W)
    print(i,': ',before_loss,loss()) 
    if before_loss-loss() < epsilon:
        break 
    # hiển thị đồ thị 
    plt.plot(X,y,'ro')
    plt.xlabel('X')
    plt.ylabel('Y')
    x0 = np.linspace(start=0, stop=10, num=50)
    y0 = W[1]*x0+W[0]
    plt.plot(x0,y0)
    plt.show()
    # print result 
    
```
