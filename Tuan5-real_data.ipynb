```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/adult.data")
df=df.sample(n=200)
```

```python
data = df.values
print(data.shape)
y = data[:, -1:]
y = y.ravel()
print(y[:3])
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
y = LE.fit_transform(y)
print(y[:10])
y1 = y[y==1]
y0 = y[y==0]
print(y1)
print(y0)
```

```python
dataX = df.select_dtypes(include=['float64', 'int64'])
print(dataX.shape)
print(dataX)
x = dataX.values
print(x)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)
X = scaler.transform(x)
print(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=4)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn import svm

svm_classifier = svm.SVC(kernel='linear', gamma=0.01, C=100.)
svm_classifier.fit(X_train, y_train)

y_pred_svm = svm_classifier.predict(X_test)
print('predict:\n', y_pred_svm)
print('y_true:\n', y_test)

acc_score_svm = metrics.accuracy_score(y_pred_svm,y_test)
print('acurracy:\n', acc_score_svm)
```
